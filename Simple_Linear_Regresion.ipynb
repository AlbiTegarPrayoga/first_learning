```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/drive/MyDrive/p2d/data-SLR.csv')

print(df)
```

```python
f = plt.figure(figsize=(12,4))
f.add_subplot(1,2,1)
df['Income'].plot(kind='kde')
f.add_subplot(1,2,2)
plt.boxplot(df['Income'])
plt.show()
```

```python
f = plt.figure(figsize=(12,4))
f.add_subplot(1,2,1)
df['Consumption'].plot(kind='kde', c='g')
f.add_subplot(1,2,2)
plt.boxplot(df['Consumption'])
plt.show()
```

```python
df.corr()
```

```python
df.head()
```

```python
x = df['Income'].values.reshape(-1,1)
y = df['Consumption'].values.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(x_train, y_train)
```

```python
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
lin_reg.score(x_test, y_test)
```

```python
y_prediksi = lin_reg.predict(x_test)
plt.scatter(x_test, y_test)
plt.plot(x_test, y_prediksi, c='r')
plt.xlabel('Income')
plt.ylabel('Consumption')
plt.title('Plot Income vs Consumption')
```
